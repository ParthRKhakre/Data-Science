# Introduction to API


> API stands for Application Programming Interface.
> API is the connector between two interface.

# Why do we need APIs


# Pre API Era
Lets consider a scenario where No API is present used just frontend interacting with the backend and backend access DB
without API the website is created as a single application
i.e. frontend and backend are tightly coupled this is known as monolithic architecture.

1. Drawback of the architecture is that any bug affects the entire system.  
2. Also if any third party wants to use the services of the application then we need to provide the access to the DB or backend to third party it is not safe 



These problems are solved by the API
1. Decoupling of frontend and backend 
2. provide Third party services without providing direct access to the backend 


> Communication of the 3rd party happens with API and Backend via https Protocol and response from the backend is in the form of a JSON(Universal data format)

# Why JSON
Beacuse it is a universal data form as all languages return response in JSON format.

# Frontend Backend Interaction 
Interactions of multiple platform can use a single backend using the same API over the internet.





All the development was done in single file monolithic architecture 

# API to overcome the Mono-Architecture


In ML we have a ML model in backend interaction instead of a DB as we have in API.

# What is FAST API


FAST API is created over libraries Starlette and Pydantic


1. Older frameworks have more latency and low performance it is not good for industry level application 
2. Code writing was ugly as alot of boilerplate code was required.

# Working of API
ML model API we created a /predict endpoint it takes feature1 and feature2 as the input and provide the prediction deploy it over AWS Cloud 

2 things are put on the AWS:
1. API CODE
2. Web Server - using the ports on the machine it trys to listen the requests



> Using the endpoint User Input are transmitted using the http request 

 

> The above http request is provided to the web server, web server helps sending the request to API Code and API code will extract the features and generate the predictions 



> In above protocol http request are used but python doesn't understand the http hence we need to convert http content to a format that is understanded by the python to do this we use SGI (Server Gateway Interface)

# Case of Flask


* WSGI stands for WebSGI used in Flask. 
* WSGI is Synchronous in nature.
* Flask use Werkzeug library for WSGI purpose.

# Case of FAST API 


* ASGI used in FAST API.
* ASGI is asynchoronus in nature.
* FAST API uses Starlette library for ASGI Implementation.
