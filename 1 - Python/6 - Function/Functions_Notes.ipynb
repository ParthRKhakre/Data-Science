# Types of arguments
* Default Argument
* Positional Argument
* Keyword Argument

```python
# Default Argument : 
# while implementing a function we need to preset some values 
# to input variable inorder to avoid function failures.
# Suppose you created a function that takes two input and user 
# Enters only one input in such scenarios function can create error.

def power(a = 1,b = 1): # a,b are parameter 
    return a**b

power(9) # 9 is a argument
```

```python
# Positional Argument :
# the sequence of the argument is same as provided in function definition 
power(2,3)
```

```python
# Keyword Argument : 
# to avoid the positional argument confusion we can use keyword argument
# Precedence of Keyword argument is greater than Positional 

power(b=3,a=2)
```

# *args and **kwargs
are the special Python keywords that are used to pass the variable length of arguments to a function 


```python
# *args
# allows us to pass a variable number of non-keyword arguments to a function.
# When its required to take variable argument input to the function
# args can be used

# *args - tuple 
 

def multiply(*args): 
# also args is not a keyword you can use any word along *arguments_variable
    output = 1
    for i in args:
        output = i * output
    
    print(args) # args are the tuples that store multiple argument
    return output    
```

```python
multiply(1,2,3,4,5,6,7,8,9)
```

```python
# Accessing Documentation of any function can be done as
# function_name.__doc__
print(print.__doc__)
```

```python
# **kwargs(Keyword argument)
# allows us to pass any number of keyword arguments
# keyword arguments mean that they contain a key-value pair like dictionary in python

# **kwargs - dictionary

def display(**kwargs):
    # anyword can be used after**
    for(key,value) in kwargs.items():
        print(key,'->',value)
        
    print(kwargs)    
```

```python
display(india='delhi',srilanka="colombo",nepal="Kathmandu")
```

# Global keyword

```python
# Changing the global variable from local scope is possible 
# you need to you the global keyword 

def i(y):
    global x # it allow the global to be changed using local scope
    x += 1
    
x = 5
i(x)
print(5)    

# Everything is a pass by reference in python
```

# Lambda Functions
A lambda function is a small anonymous functions.

A lambda function can take any number of arguments, but can only have one have one expression.



```python
# x -> x**2
x = lambda x:x**2
x(2)
```

```python
# x,y -> x + y
y = lambda x,y: x+y

y(10,9)
```

# Difference between lambda VS Normal function

* No Name
* lambda has no return value 
* lambda is written in 1 line 
* not reusable 

Then why use lambda functions?
* They are used with high order functions 

```python
# check if a string has 'a'
a = lambda s: 'a' in s
a('hello')
```

```python
# Odd or Even 
b = lambda x: 'even' if x % 2 == 0 else 'odd'
b(9)
```

# Map() function 
It expects two things as an input lambda functions and iterable(List).

You can apply any type of logic on the list 

```python
# square the items in the lists 
list(map(lambda x:x**2,[1,2,3,4,5]))
```

```python
# Odd/Even labelling of the list items 
L = [1,2,3,4,5]
list(map(lambda x:'even' if x%2 == 0 else 'odd',L))
```

```python
# fetch name from a list of dict

users = [
    {
        'name':'Rahul',
        'age':49,
        'gender':'male'
    },{
        'name':'Nitish',
        'age':42,
        'gender':'male'
    },{
        'name':'Ankita',
        'age':40,
        'gender':'female'
    }
]

list(map(lambda users:users['name'],users))
```

# Filter() 
* is a built-in function in Python used to select elements from an iterable based on a condition.
* filter(function, iterable)


```python
# Numbers greater than 5 
L = [1,2,3,4,5,6,7,8]
list(filter(lambda x: x > 5,L))
```

```python
# fetch fruits starting with 'a'
L = ['apple','pine','onion','orange']
list(filter(lambda x: x.startswith('a'),L))
```

# Reduce()
* in Python is used to reduce an iterable to a single value by repeatedly applying a function.
* It is not a default functions it is present inside a module name functools
> from functools import reduce
> 

```python
# Add all elements from the list 
from functools import reduce
L = [1,2,3,4,5,6]

reduce(lambda x,y:x+y,L)
```

```python
# Find min
L = [1,21,0,13,-1]

# reduce(lambda x,y:x if x < y else y,L)
reduce(lambda x,y:min(x,y),L)
```
