# Reading Images

```python
import cv2 as cv

img = cv.imread('image/cat.jpg')
# cv.imread('path') is a function used to load an image from 
# a specified path into memory. It returns a numpy array that 
# can be used for image processing, analysis. 
```

```python
cv.imshow('Cat',img)
# imshow() is used to display an image in a window 

cv.waitKey()
# waitKey() is a function used to display a window for a given millisecond 
# until any key is pressed 
# if 0 is passed in argument it waits till any key is pressed
# if e.g passed 5000 the window is open for 5 second

cv.destroyAllWindows()
# remove all the window from the console 
```
