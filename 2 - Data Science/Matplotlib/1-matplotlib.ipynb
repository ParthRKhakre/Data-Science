# 2D Line Plot
* Bivariate Analysis - Two Column selected from dataset
* Can be applied on :
1) Both Columns are Numerical
2) One Column is Numerical and Other is Categorical
* Use Case of 2D line Plot is Time Series

```python
# plotting simple function 
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

# plt.plot(x-axis,y-axis)
plt.plot(year,price)
```

```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman.head()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plot 
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title("Most Runs")
plt.xlabel("Years")
plt.ylabel("Runs")
```

```python
# color(hex) and line(width and style) and market
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue')

plt.title('Virat and Rohit Comparison')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
# types of linestyle : dashed,dotted,dashdot,solid
# line width : sets the width of the line 
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashed')

plt.title('Rohit Sharma vs Virat Kohli Carrier Comparison')
plt.xlabel('Session')
plt.ylabel('Runs Scored')
```

```python
# marker = +,D,.,>,o , a marker is a intersection point
# markersize - size of marker
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted',linewidth=1.2,marker='D',markersize=2)
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashdot',linewidth=3,marker='+',markersize=8)

plt.title('Rohit Sharma vs Virat Kohli Carrier Comparison')
plt.xlabel('Session')
plt.ylabel('Runs Scored')
```

```python
# legend() - function is used to create a legend 
# on the graph to provide more information.

plt.plot(batsman['index'],batsman['V Kohli'],linewidth=3,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashed',label='Rohit')

plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend(loc='best') 

# by setting the loc parameter we can put the legend into a certain position
# loc = best,upper right,upper left,right,center left,center right,etc
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000] 
# here 4500000 is a outlier due to which graph gets disturbed hence trimming is required 
# trimming can be done using ylim() function give range before the outlier
years = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(years,price)

# trimming x,y axis
plt.ylim(0,100000)
plt.xlim(2016,2020)
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],linewidth=3,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashed',label='Rohit',marker='D')

plt.xlabel('Season')
plt.ylabel('Runs')
plt.grid()

# In any IDE where you  are working on the graph plt.show() 
# plt.show() is used to display graph on screen

plt.show()
```
